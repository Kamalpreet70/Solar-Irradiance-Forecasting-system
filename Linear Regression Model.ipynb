# Cold season

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np
```

Loading and splitting data


```python
# Load the data
data = pd.read_csv('cold_season_NITJ_data.csv')

# Split the data into features (X) and target variable (Y)
X = data[['Temperature', 'Pressure', 'Relative Humidity', 'Solar Zenith Angle',
          'Wind Speed', 'Wind Direction', 'Dew Point', 'Precipitable Water',
          'Clearness Index']]
Y = data['GHI']
```

Splitting data into training, validation, testing sets

```python
# Split the data into training, testing, and validation sets
X_train, X_temp, Y_train, Y_temp = train_test_split(X, Y, test_size=0.3, random_state=42)


# Further split training data into training and validation sets (70% train, 30% validation)
X_test, X_val, Y_test, Y_val = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)
```

Model fittinh

```python
# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)
```

Making predictions

```python
# Make predictions on the testing set
Y_pred = model.predict(X_test)
```

Metrics

```python
# Calculate evaluation metrics
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R-squared (R2):", r2)
print("Mean Absolute Error (MAE):", mae)
```
